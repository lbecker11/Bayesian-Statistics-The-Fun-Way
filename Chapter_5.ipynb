1. You want to use the beta distribution to determine whether or not a coin you have is a fair one - meaning P(heads) and P(tails) is equal. You flip it 10 times and get 4 heads and 6 tails. Using the beta distribution, what is the chance that the coin will land on heads more than 60% of the time?
 - Beta distribution is a continous motion of binominal PMF, where we change the *p* instead of *k* to find the likeliness of a *p*
 - $\text{Beta(p;}\alpha,\beta)=\frac{p^{\alpha-1}\times (1-p)^{\beta-1}}{beta(\alpha, \beta)}\text{where }\alpha \text{ is how many times we observe an event we care about, }\beta \text{ is how often the event we are about didnt happen, and p the probaility of an event.}$
 - $beta(\alpha, \beta)\text{ is the integral of p between every value of 0 to 1 (or defined) for }p^{\alpha-1}\times (1-p)^{\beta-1}$
 - this is the probability density function (PDF)

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
y = beta.pdf(x, 4, 6)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 0.6) & (x <= 1),color='lightgreen', alpha=0.5, label='Shaded Area')
plt.axvline(0.6, color='red', linestyle='--', linewidth=2)
plt.ylabel('Density')
plt.xlabel('p')
plt.show()
print(f'There is about a {beta.sf(0.6, 4, 6)*100:.2f}% chance that the true probability of getting heads is 60% or greater')
#beta.sf (survival function) tells us the area under the curve for x > p 
#beta.cdf (cumulative distribution function) tells us the area under the curve for x < p
```

2. You flip the coin 10 more times and now have 9 heads and 11 tails. What is the *p* that the coin is fair give or take 5%?

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0, 1, 100)
alpha = 9
beta_ = 11
y = beta.pdf(p, alpha, beta_)

plt.plot(p, y)
plt.fill_between(p, y, where=(x>=0.5) & (x<=1), color='lightgreen', alpha=0.5, label='Shaded Area')
plt.axvline(0.5, color='red', linestyle='--', linewidth=2)
plt.ylabel('Density')
plt.xlabel('p')
plt.show()
print(f'The P of the coin being fair based of the observed events is: {(beta.sf(0.45, alpha, beta_)-beta.sf(0.55, alpha, beta_))*100:.2f}%')
```
